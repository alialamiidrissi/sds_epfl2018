```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
A = (np.random.uniform(-1, 1, size=2000)).reshape(1000,2)
```

```python
plt.scatter(A[:,0], A[:,1])
```

```python
labels = []

for i in A:
    if(i[1] > 0):
        labels.append(1)
    else:
        labels.append(-1)
```

```python
file1 = open("toy_data.txt", "w") 

for i, data in enumerate(A):
    file1.write(str(i) + " 0:" + str(data[0]) + " 1:" + str(data[1]) + "\n") 
file1.close()
    
file2 = open("toy_labels.txt","w")  
for label in labels:
    file2.write(str(label) + "\n") 
file2.close()
```
