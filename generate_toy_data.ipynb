```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
A = (np.random.uniform(-1, 1, size=2000)).reshape(1000,2)*1000
```

```python
plt.scatter(A[:,0], A[:,1])
```

```python
labels = []

for i in A:
    if(i[1] > 0):
        labels.append(1)
    else:
        labels.append(-1)
```

```python
file1 = open("./data/toy_data.txt", "w") 

for i, data in enumerate(A):
    file1.write(str(i) + " 0:" + str(data[0]) + " 1:" + str(data[1]) + "\n") 
file1.close()
    
file2 = open("./data/toy_labels.txt","w")  
for label in labels:
    file2.write(str(label) + "\n") 
file2.close()
```

```python
labels_a,labels_b = [],[]
for i in A:
    if(i[1] > 0):
        labels_a.append(i)
    else:
        labels_b.append(i)
labels_a= np.array(labels_a)
labels_b= np.array(labels_b)
```

```python
plt.scatter(labels_a[:,0], labels_a[:,1],color='r')
plt.scatter(labels_b[:,0], labels_b[:,1],color='g')
```
